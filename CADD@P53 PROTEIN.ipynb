```python
# Example 1: Loading and Anaysing Protein structures(PDB Files)
# Task: Use Biopython to fetch and analyse a protein structure from the PDB database

from Bio.PDB import PDBList, MMCIFParser

pdbl = PDBList()

pdbl.retrieve_pdb_file("4RP6", pdir=".", file_format="mmCif")

parser =MMCIFParser()

# The retrieve_pdb_file function downloads the file to a subdirectory based on the PDB ID.
# The path should be ./<PDB ID>/<PDB ID>.cif
Structure = parser.get_structure("4RP6", "./4rp6.cif")

print(f"Models: {len(Structure)}, Chains: {len(Structure[0])}")
```

```python
# Example 3: Analysing DNA/Protein Sequences (FASTA files)
# Task: Read a FASTA file and compute GC content

from Bio import SeqIO
from Bio.SeqUtils import gc_fraction

# Read a FASTA file and iterate through records

for record in SeqIO.parse("/content/rcsb_pdb_4RP6.fasta", "fasta"):
    gc_content = gc_fraction(record.seq) * 100

    # Convert fraction into percentage

    print(f"GC Content for record {record.id}: {gc_content:.2f}%")
```
